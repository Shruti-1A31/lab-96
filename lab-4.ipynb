```python
#Using universal function.
b = np.array([5,10,15,20,25])
print("Original array: ",b)

print("\nnp.add(b, 2):",np.add(b, 2))
print("\nnp.subtract(b, 5):", np.subtract(b, 5))
print("\nnp.multiply(b, 5):", np.multiply(b, 5))
print("\nnp.divide(b, 2):", np.divide(b, 2))
```

```python
#Using Broadcasting.
c = np.array([[5],[9]])
c1 = np.array([2,4])
print("Original array c: \n",c)
print("Original array c1: \n",c1)

print("\nnp.add(c,c1): \n",np.add(c,c1))
print("\nnp.subtract(c,c1): \n", np.subtract(c,c1))
print("\nnp.multiply(c,c1): \n", np.multiply(c,c1))
print("\nnp.divide(c,c1): \n", np.divide(c,c1))
```

```python
''' Rathod Shruti A.
     216270307096    '''

'''Write a Pandas program to implement following operation  to sort the DataFrame first 
by 'name' in ascending order '''

data_new = data.sort_values(by = ['sname'], ascending=True)
print(data_new)
```

```python
''' Rathod Shruti A.
     216270307096    '''

'''Write a Pandas program to implement following operation to delete the one specific
column from the DataFrame'''

del data['sem']
print(data)
```
